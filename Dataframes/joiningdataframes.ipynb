```python
import pandas as pd
```

```python
# you can join DataFrames with the merge() method
# there are 4 primary type of joins inner left right and outer
# inner and left joins are the most common, right is rarely used
```

```python
#inner returns record that exist in both tables -how="inner"
#left returns all records from the left table and any matching records from the right table -how="left"
#outer returns all records from both tables, including non matching records
```

```python
#.join() method can join two dataframes based on their index
#can also join based on indices with .merge() but specify left_index=True and right_index=True
```

```python
transactions = pd.read_csv("../retail/transactions.csv", parse_dates=["date"])

oil = pd.read_csv("../retail/transactions.csv", parse_dates=["date"])
```

```python
transactions.info()
```

```python
oil.info()
```

```python
transactions.merge(oil,
                   how="inner",
                   left_on=["date"],
                   right_on=["date"]
                  )
```

```python
transactions.merge(oil,
                   how="left",
                   left_on=["date"],
                   right_on=["date"]
                   )
```
